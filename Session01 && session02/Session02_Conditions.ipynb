```python
## limit : score >= 0 and score <= 20 ===> out of range
## score >= 17 ===> A++
## score < 17 and score >= 15 ===> A+
## score < 15 and score >= 10 ===> A
## score < 10 ====> B
```

```python
score = 12
if score < 0 or score > 20:
  print("out of range!")
elif score >= 17:
  print("A++")
elif score >= 15:
  print("A+")
elif score >= 10:
  print("A")
else:
  print("B")
```

```python
## FizzBazz
## number % 2 == 0 ====> F
## number % 3 == 0 ===> B
## number % 2 == 0 and number % 3 == 0 ===> FB
## else ======> number
```

```python
number = int(input("your number: ")) # number = 12
if number % 2 == 0 and number % 3 == 0:
  print("FB")
elif number % 2 == 0: # 12 % 2 == 0 ====> True
  print("F")
elif number % 3 == 0:
  print("B")
else:
  print(number)
```
