### copy (shallow copy)

```python
x = 10
y = x
print(f"x value:{x}, y value: {y}")
x = 20
print(f"x value:{x}, y value: {y}")
```

```python
list1 = [2, 17, 13]
list2 = list1
print(f"list1:{list1}, list2: {list2}")
list1.append(50)
print(f"list1:{list1}, list2: {list2}")
```

```python
list1 = [3, 17, 18]
list2 = list1[:]
print(f"list1:{list1}, list2: {list2}")
list2.append(50)
print(f"list1:{list1}, list2: {list2}")
```

```python
list1 = [3, 17, 18]
list2 = list(list1)
print(f"list1:{list1}, list2: {list2}")
list2.append(50)
print(f"list1:{list1}, list2: {list2}")
```

```python
list1 = [3, 17, 18]
list2 = list1.copy()
print(f"list1:{list1}, list2: {list2}")
list2.append(50)
print(f"list1:{list1}, list2: {list2}")
```

### Points

```python
num = int(input("number of your students: "))
scores = []
for i in range(num):
  score = float(input(f"student_{i+1}: "))
  scores.append(score)
print(scores)
min_score = min(scores)
print(min_score)
max_score = max(scores)
print(max_score)
sum_score = sum(scores)
print(sum_score)
avg = sum_score / len(scores)
print(avg)
```
